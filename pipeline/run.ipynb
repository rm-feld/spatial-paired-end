```python
_mutlengths = []
_num_reads_by_cov = []
_num_copies = []
_signal = []
_peaks = []
_locs = []
_fnames = []
_times = []

coverages = np.linspace(1, 10, 5)
for j in range(len(coverages)):
    mutlengths = []
    num_reads_by_cov = []
    num_copies = []
    signal = []
    peaks = []
    locs = []
    fnames = []
    times = []
    coverage = coverages[j] 
    for i in range(50):
        start = time.time()
        fp, aug_fp, tm, root, celldict, focal = m.genstep(seq_dict, _seqdir, log = log,
                                                    easy = True, only_insertions=True)
        print(tm)
        is_insertion = (tm.num_copies != 0)
        length = tm.length
        rl = tm.refstart 
        rr = rl + length

        mutlengths.append(length)
        locs.append((rl, rr))
        num_copies.append(tm.num_copies)
        print(f"num copies: {num_copies[-1]}")
        fnames.append(aug_fp)

        num_reads_by_coverage_mult = coverage * tm.length * tm.num_copies // 300 
        read_dict["numreads"] = int(num_reads_by_coverage_mult)
        num_reads_by_cov.append(num_reads_by_coverage_mult)

        readdir = f"{_readdir}/b{i}"
        subprocess.call(["mkdir", "-p", readdir])
        
        try:
            df, sus_df = m.readstep(read_dict, readdir, f"b{i}", tm, aug_fp, log = log)
            signal.append(len(df[((df["l"] >= rl) & (df["r"] <= rr)) & (df["l"] > df["r"])]))
            peaks.append(len(df[df["l"] > df["r"]]))
        except:
            peaks.append(-1)
            signal.append(-1)
        
        times.append(time.time() - start)
    
    _mutlengths.append(mutlengths)
    _num_reads_by_cov.append(num_reads_by_cov)
    _num_copies.append(_num_copies)
    _signal.append(signal)
    _peaks.append(peaks)
    _locs.append(locs)
    _fnames.append(fnames)
    _times.append(times)
```
